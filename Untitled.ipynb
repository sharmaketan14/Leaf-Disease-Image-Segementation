# Resizing, Rotation & Cropping

```python
img_4 = cv.imread("88112cattura-1.jpg")
cv.imshow("Original",img_4)
cv.waitKey(0)
cv.destroyAllWindows()
image_re_linear = cv.resize(img_4, None, fx=5.5, fy=5.5, interpolation=cv.INTER_LINEAR)
image_re_cubic = cv.resize(img_4, None, fx=5.5, fy=5.5, interpolation=cv.INTER_CUBIC)
cv.imshow("Linear", image_re_linear)
cv.waitKey(0)
cv.destroyAllWindows()
cv.imshow("Cubic", image_re_cubic)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
height, width = img_4.shape[:2]
rotation_matrix = cv.getRotationMatrix2D((width/2, height/2), 10, 1)
rotation_img = cv.warpAffine(img_4, rotation_matrix, (width, height))
cv.imshow("Rotated", rotation_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img_cropped = img_4[0:900, 300:540]
cv.imshow("Cropped", img_cropped)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Hough Line Detection

```python
image = cv2.imread("corner.jpg")
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
```

```python
edges = cv2.Canny(image_gray, 50, 150, apertureSize=3)
lines = cv2.HoughLines(edges, 1, np.pi/180, 200)

for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 1000*(-b))
    y1 = int(y0 + 1000*(a))
    x2 = int(x0 - 1000*(-b))
    y2 = int(y0 - 1000*(a))
    
    cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 2)
plt.imshow(image)
```
